```python
sns.barplot(x='SepalLengthCm', y='Species', data= df_iris);
```

```python
#utilizando o numpy para criar uma coluna nova com uma resposta de uma duvida
import numpy as np
df_iris['PETALA GRANDE/PEQ'] = np.where(df_iris['PetalLengthCm'] >= 6, 'Grande', 'Pequena')
df_iris
```

```python
#criando um grafico com a nova coluna como um novo parametro
sns.barplot(y='SepalLengthCm',x='Species', hue= 'PETALA GRANDE/PEQ', data= df_iris);
```

```python
#criando função para filtro
def low_valor_red(value):
    color = 'red' if value < 92 else 'White'
    return f'color: {color}'

fifa[['overall','potential']].head(10).style.map(low_valor_red)
```

```python
#Aplicando o filtro em toda a tabela
fifa.head(10).style.map(low_valor_red, subset = ['overall', 'potential'])
```

```python
#Função para mudar a cor do fundo da cedula que seja a maior da colulna
def maiores_da_coluna(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
#Aplicando em toda a tabela
fifa.head(10).style.apply(maiores_da_coluna, subset= ['overall','age','potential'])
```

Mais sobre filtros !

```python
fifa.head()
```

```python
#usar o .loc para selecionar um intervalo de linhas, porem mostrando apenas a coluna que foi digitada apos a virgula
fifa.loc[0:5, 'age']
```

```python
#para usar o .iloc nao se usa o nome para filtro, sim um intervalo de colunas
fifa.iloc[0:4, 2:5]
```

```python
#filtrar com criterios;
fifa.loc[(fifa['age'] < 24) & (fifa['overall'] > 85)]
```

```python
#Filtro de comparação
fifa[fifa['lcb']==fifa['rcb']]
```

```python
#Utilizando a função .filter( )
fifa.filter(['age','potential','short_name'])
```

```python
#Criando a media dos overall e filtazendo uma query 
fifa_median = fifa['overall'].median()
fifa.query('overall > @fifa_median')
```

```python
#.isin( ) para saber se tal agurmento esta no lugar pedido, e usado para criar um novo df
champs = fifa[fifa['nationality_name'].isin(['Brazil','Argentina','France','England','Spain','Italy','Uruguay','Germany'])]
champs.head()
```

```python
# se colocado o acento ~ antes dos agurmentos o codigo passa a mostrar coisas que nao estao dentro do paramentro
non_champs = fifa[~fifa['nationality_name'].isin(['Brazil','Argentina','France','England','Spain','Italy','Uruguay','Germany'])]
non_champs.head()
```

```python
#Selecionando apenas numericos do dataframe
numerics = ['int16','int32','int64','float16','float32','float64']
fifa_non_numerics = fifa.select_dtypes(include=numerics)
fifa_non_numerics
```

```python
#Selecionando os não numericos do data frame
numerics = ['int16','int32','int64','float16','float32','float64']
fifa_non_numerics = fifa.select_dtypes(exclude= numerics)
fifa_non_numerics
```

```python
#Ordenando por argumentos
fifa.sort_values(by='potential', ascending= True)
```

```python
#selecinando colunas desejadas, e quais estaticas das mesmas mostrar
fifa.agg({'potential': ['min','mean','max'],
          'age': ['min','mean','max']})
```

```python
#Agrupando uma coluna por 2 argumentos e mostrando as medias.
champs.groupby('nationality_name')[['overall','age']].mean().sort_values(by='overall')
```
