```python
#criando função para filtro
def low_valor_red(value):
    color = 'red' if value < 92 else 'White'
    return f'color: {color}'

fifa[['overall','potential']].head(10).style.map(low_valor_red)
```

```python
#Aplicando o filtro em toda a tabela
fifa.head(10).style.map(low_valor_red, subset = ['overall', 'potential'])
```

```python
#Função para mudar a cor do fundo da cedula que seja a maior da colulna
def maiores_da_coluna(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
#Aplicando em toda a tabela
fifa.head(10).style.apply(maiores_da_coluna, subset= ['overall','age','potential'])
```

```python
#importando um novo data frame e a biblioteca ploty
import plotly.offline as py
import plotly.graph_objs as go

df_plo = pd.read_csv('marketing_campaign.csv', sep = '\t')
df_plo.head(5)
```

```python
#tratando as datas
df_plo['Dt_Customer'] = pd.to_datetime(df_plo['Dt_Customer'], dayfirst=True)

#criando as novas colunas
df_plo['Year_Customer']= df_plo['Dt_Customer'].dt.year
df_plo['Month_Customer']= df_plo['Dt_Customer'].dt.month
df_plo['Day_Customer']= df_plo['Dt_Customer'].dt.day
```

```python
#criando series onde o index é o mes e o valor é avg_income
data = df_plo[df_plo['Year_Customer'] == 2013].groupby('Month_Customer')['Income'].mean()

#definindo as variaveis
month = data.index
avg_income = data.values
```

```python
line1 = go.Scatter(x = month, y = avg_income, mode = 'lines', name = 'Ligado por linhas')

layout = go.Layout(title= 'Salario Médio dos clientes que entraram nos meses de 2013',
                   xaxis = {'title': 'Mês'},
                   yaxis = {'title': 'Salário Médio'})

fig = go.Figure(data= [line1], layout= layout)

py.iplot(fig)
```

```python
line1 = go.Scatter(x = month, y = avg_income, mode = 'markers', name = 'Ligado por linhas')

layout = go.Layout(title= 'Salario Médio dos clientes que entraram nos meses de 2013',
                   xaxis = {'title': 'Mês'},
                   yaxis = {'title': 'Salário Médio'})

fig = go.Figure(data= [line1], layout= layout)

py.iplot(fig)
```

```python
#Função para criar graficos de barra simples
df_iris_top10= df_iris.head(9)

def grafico(x,y,data,legx,legy):
    plt.figure(figsize= (10,5))
    plt.bar(x,y, data=data,linewidth=4, color = 'b')
    plt.legend()
    plt.xlabel(legx)
    plt.ylabel(legy)
    plt.xticks(rotation = 90)
    plt.tight_layout()
    plt.show()

    return grafico



grafico(fifa10['short_name'],fifa10['overall'],fifa10,'Jogadores','Overall')
grafico(df_iris_top10['Id'],df_iris_top10['PetalLengthCm'], df_iris_top10,'Species','Tam. Petala')
```

### Utilizando o Pydantic

```python
from pydantic import BaseModel,Field, ValidationError
from typing import List, Optional

class User(BaseModel):
    id: int
    name: str
    email: str
    age: Optional[int] = None # Campo opcional, se nao conter = vazio
    is_active: bool = True
    tags: List[str] = [] #lista de string por padrao vazia

data_valid = {
    "id": 123,
    "name": "Alice Smith",
    "email": "alice@example.com",
    "age": 30,
    "is_active": True,
    "tags": ["developer", "python"]
}

user1 = User(**data_valid) #criar estancia apartir do dicionario
print('Usuario 1 (valido):\n',user1)
print(f'Nome: {user1.name},email: {user1.email}')

print('-' * 30)

#Dados com campo opcional e padrao

dataminimo = {
    'id': 32,
    'name': 'Jason',
    'email': 'jason@teste.com'
}

user2 =User(**dataminimo)
print('Usuario 2 (valido):',user2)
print(f'Nome:{user2.name}, idade: {user2.age}')

print('-' * 30)

#dados invalidos ou 'errados'
data_invalid = {
    'id': 'Abc',
    'name': 'Jose antonio',
    #sem email
    'age': 25
}

try:
    user3 = User(**data_invalid)
except ValidationError as e:
    print('erro de digitação no usuario 3:')
    print(e.json())
```
