*Graficos com seaborn*

```python
#importando o seaborn
import seaborn as sns
```

```python
#importando um novo dataframe
df_iris = pd.read_csv('Iris.csv')
df_iris.head()
```

```python
#Grafico reletional plot
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=(df_iris))
```

```python
#adicionando o hue, para separar por cor
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=(df_iris), hue='Species')
```

```python
#adicionando o style, para mostrar no grafico as diferenças
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=(df_iris), hue='Species', style='Species')
```

```python
#Criando um grafico de barras basico
sns.barplot(y='SepalLengthCm', x='Species', data= df_iris);
```

```python
sns.barplot(x='SepalLengthCm', y='Species', data= df_iris);
```

```python
#utilizando o numpy para criar uma coluna nova com uma resposta de uma duvida
import numpy as np
df_iris['PETALA GRANDE/PEQ'] = np.where(df_iris['PetalLengthCm'] >= 6, 'Grande', 'Pequena')
df_iris
```

```python
#criando um grafico com a nova coluna como um novo parametro
sns.barplot(y='SepalLengthCm',x='Species', hue= 'PETALA GRANDE/PEQ', data= df_iris);
```

```python
#criando função para filtro
def low_valor_red(value):
    color = 'red' if value < 92 else 'White'
    return f'color: {color}'

fifa[['overall','potential']].head(10).style.map(low_valor_red)
```

```python
#Aplicando o filtro em toda a tabela
fifa.head(10).style.map(low_valor_red, subset = ['overall', 'potential'])
```
