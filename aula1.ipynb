```python
#Função para mudar a cor do fundo da cedula que seja a maior da colulna
def maiores_da_coluna(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max]
```

```python
#importando um novo data frame e a biblioteca ploty
import plotly.offline as py
import plotly.graph_objs as go

df_plo = pd.read_csv('marketing_campaign.csv', sep = '\t')
df_plo.head(5)
```

```python
#tratando as datas
df_plo['Dt_Customer'] = pd.to_datetime(df_plo['Dt_Customer'], dayfirst=True)

#criando as novas colunas
df_plo['Year_Customer']= df_plo['Dt_Customer'].dt.year
df_plo['Month_Customer']= df_plo['Dt_Customer'].dt.month
df_plo['Day_Customer']= df_plo['Dt_Customer'].dt.day
```

### Utilizando o Pydantic

```python
from pydantic import BaseModel,Field, ValidationError
from typing import List, Optional

class User(BaseModel):
    id: int
    name: str
    email: str
    age: Optional[int] = None # Campo opcional, se nao conter = vazio
    is_active: bool = True
    tags: List[str] = [] #lista de string por padrao vazia

data_valid = {
    "id": 123,
    "name": "Alice Smith",
    "email": "alice@example.com",
    "age": 30,
    "is_active": True,
    "tags": ["developer", "python"]
}

user1 = User(**data_valid) #criar estancia apartir do dicionario
print('Usuario 1 (valido):\n',user1)
print(f'Nome: {user1.name},email: {user1.email}')

print('-' * 30)

#Dados com campo opcional e padrao

dataminimo = {
    'id': 32,
    'name': 'Jason',
    'email': 'jason@teste.com'
}

user2 =User(**dataminimo)
print('Usuario 2 (valido):',user2)
print(f'Nome:{user2.name}, idade: {user2.age}')

print('-' * 30)

#dados invalidos ou 'errados'
data_invalid = {
    'id': 'Abc',
    'name': 'Jose antonio',
    #sem email
    'age': 25
}

try:
    user3 = User(**data_invalid)
except ValidationError as e:
    print('erro de digitação no usuario 3:')
    print(e.json())

print('-'*30)

```

```python
#Validação aninada

from pydantic import BaseModel,Field, ValidationError, EmailStr
from typing import List, Optional


class endereco(BaseModel):
    rua: str
    numero: int
    estado: str
    cep: str

class empresa(BaseModel):
    nome: str
    nome_ceo: Optional[str] = None # opcional se nao conter == None
    endereco: endereco #Campo aninado

class funcionario(BaseModel):
    id: int
    nome: str
    cargo: str
    email: EmailStr
    skils: list[str] = []
    empresa: empresa

data_valid1 = {
    'id': 123,
    'nome': 'Lucas',
    'cargo': 'Analista',
    'email': 'lucas@teste.com',
    'skils':['numpy','seaborn'],
    'empresa': {
        'nome': 'Analise e dados',
        'nome_ceo': 'Roberto',
        'endereco':{
            'rua': 'rua 1',
            'estado':'São Paulo',
            'numero': 123,
            'cep': '1174-550'
        }
    }
}
try:
    dadosv = funcionario(**data_valid1)
    print('dados validos do funcionario 1:',dadosv)
except ValidationError as e:
    print('erro de validação do funcionario 1')
    print(e.json())


```

```python
#criar contagem
smokes_counts = df_saude['smoker'].value_counts()

#tamanho do grafico
plt.figure(figsize=(8,6))

#criar gafico de barras com indece e contagem
plt.bar(
    smokes_counts.index, #valor do eixo x
    smokes_counts.values, #valor do eixo y
    color=['steelblue','lightcoral'] #cores das barras
)

#rotulos deo eixo x, letra tamnho 8
plt.xlabel('Smoker', fontsize=8)

#rotulo do eixo y, lettra tamanho 8
plt.ylabel('Count', fontsize=8)

#Titulo, letra tamanho 14
plt.title('Distribuição de Fumantes', fontsize= 14)

plt.show()
```

Veja que o gráfico no Matplotlib é bastante intuitivo, você apenas vai construindo parte por parte dele e adicionando (ou removendo) o que deseja através de comandos específicos, como xlabel (rótulo do eixo x), ylabel (rótulo do eixo y) e title (título do gráfico)!
