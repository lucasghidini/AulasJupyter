```python
import pandas as pd
import numpy as np
```

```python
studs = pd.read_csv(r'C:\Users\Lucas\Desktop\Projetos\Treinamento jupyter\DatasSets_pandas\data.csv', sep=';')
studs.head()
```

```python
#Proporção de dropuout
studs['Target'].value_counts(1)
```

```python
#Media de inflaçao para cada tipo de target
studs.groupby('Target')['Inflation rate'].mean()
```

```python
#pivot_table
studs.pivot_table(index='Target', columns=['Marital status'],values='Inflation rate', aggfunc=np.mean)
```

```python
#cruzando mais variveis
studs.pivot_table(index='Target',
                  columns=['Daytime/evening attendance\t','Marital status'],
                  values='Previous qualification (grade)',
                  aggfunc=np.mean
                  )
```
